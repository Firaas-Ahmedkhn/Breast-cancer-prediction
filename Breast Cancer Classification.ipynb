```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score  # Calculate accuracy of a classification model
```

## Importing data

```python
cancer_data = pd.read_csv('Breast_cancer.csv')
```

```python
cancer_data.head()
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataset
```

## Loading data in panda dataset

We are loading this data in pandas data frame because the is not very structures

```python
df = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
df.head()
```

#### Loading the "target" column in data frame 

```python
df['Diagnosis'] = breast_cancer_dataset.target  # Diagnosis: 0 = benign, 1 = malignant 
```

```python
df.head()
```

```python
df['Diagnosis'].value_counts()
```

1 = Benign


0 = Malignant

```python
df.groupby('Diagnosis').mean()
```

## Separating the features and targets

```python
X = df.drop('Diagnosis', axis = 1)   # X = Input features
Y = df['Diagnosis']
```

```python
X
```

```python
Y
```

### Splitting the data into Training & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

## Model Training

Logistic Regression

```python
model = LogisticRegression()

# Training the logistic regression model 

model.fit(X_train, Y_train)
```

## Model Evaluation

Accuracy score

```python
# Accuracy on test data

X_test_prediction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print(f"The Accuracy of test data: {test_data_accuracy}")
```
